```python
#Import libary
import pandas as pd
import math
```

```python
# Load your JSONL file
df = pd.read_json("starterpacks.jsonl", lines=True)
```

```python
total = len(df)
print("Total rows:", total)

# -----------------------------
# 1. SPLIT FIRST 25,000 (5 files × 5K each)
# -----------------------------
TEST_CHUNK_SIZE = 5000
NUM_TEST_FILES = 5
TEST_TOTAL = TEST_CHUNK_SIZE * NUM_TEST_FILES  # 5000 * 5 = 25k

for i in range(NUM_TEST_FILES):
    start = i * TEST_CHUNK_SIZE
    end = start + TEST_CHUNK_SIZE
    chunk = df.iloc[start:end]
    
    filename = f"starterpacks_5K_{i+1}.jsonl"
    chunk.to_json(filename, orient="records", lines=True)
    print(f"Wrote {len(chunk)} rows to {filename}")
```

```python
# -----------------------------
# 2. REMAINING ROWS → split into 5 equal files
# -----------------------------
remaining = df.iloc[TEST_TOTAL:]

NUM_REST_FILES = 5
rest_chunk_size = math.ceil(len(remaining) / NUM_REST_FILES)

print("Each remaining file will have ~", rest_chunk_size, "rows")

for i in range(NUM_REST_FILES):
    start = i * rest_chunk_size
    end = start + rest_chunk_size
    chunk = remaining.iloc[start:end]
    
    filename = f"starterpacks_{len(chunk)}_{i+1}.jsonl"
    chunk.to_json(filename, orient="records", lines=True)
    print(f"✔ Wrote {len(chunk)} rows to {filename}")
```

```python
import math
import pandas as pd

# 1. Load your big JSONL of starter packs
df = pd.read_json("starterpacks_65939_Trang.jsonl", lines=True)

CHUNK_SIZE = 5000
total = len(df)
num_chunks = math.ceil(total / CHUNK_SIZE)

for chunk_id in range(num_chunks):
    start = chunk_id * CHUNK_SIZE
    end = min(start + CHUNK_SIZE, total)
    
    df_chunk = df.iloc[start:end]
    filename = f"Trang_chunk_{chunk_id + 1}.jsonl"
    
    df_chunk.to_json(filename, orient="records", lines=True)
    
    print(f"✔ Saved {len(df_chunk)} rows to {filename}  (rows {start}–{end-1})")
```
